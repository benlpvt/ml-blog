```python
non_religs['pct_exempt'] = (non_religs['Number of Exemptions'] / non_religs['Population Size']) * 100
non_religs.dropna(axis=0, inplace=True)
non_religs.head(20)
```

```python
from sklearn.cluster import KMeans

X = non_religs.loc[:, ['Number of Exemptions', 'pct_exempt']]
kmeans = KMeans(init='random', n_clusters=10, n_init=20, max_iter=500, random_state=42).fit(X)
```

```python
def plot_data(X):
    plt.scatter(X['Number of Exemptions'], X['pct_exempt'], c='k', s=30)

def plot_centroids(centroids, weights=None, circle_color='k', cross_color='w'):
    if weights is not None:
        centroids =  centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=35, linewidths=8,
    color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=2, linewidths=12,
    color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True, show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                        np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap='jet')
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
plt.figure(figsize=(10,7))
plot_decision_boundaries(kmeans, X)
plt.savefig('voronoi_exempts.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plot_decision_boundaries(kmeans_two, X2)
plt.savefig('voronoi_exempts_all.png')
plt.show()
```
