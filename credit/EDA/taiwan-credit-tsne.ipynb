```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import xlrd


filename = 'default of credit card clients.xls'

DATA = os.path.relpath('/content/gdrive/MyDrive/Github/ml-blog/credit/data/' + filename)

df = pd.read_excel(DATA, 'Data', index_col=[0], header=[1], na_values='NA')
df.head()
```

```python
X = df.iloc[:, :23]
Y = df.iloc[:, 23]
```

```python
from sklearn.manifold import TSNE
import plotly.express as px

X_embed = TSNE(n_components=2, learning_rate=800,
               init='pca', metric='cosine',
               perplexity=40, random_state=42).fit_transform(X)
```

```python
fig = px.scatter(
    X_embed, x=0, y=1,
    color=X.PAY_0, labels={'color': 'PAY_0'}
)

fig.show()
```
